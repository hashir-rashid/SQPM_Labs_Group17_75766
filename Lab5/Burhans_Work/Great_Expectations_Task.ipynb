##5. Set Up Great Expectations Context and Data Source

```python
import great_expectations as gx

context = gx.get_context()

# In GX 1.0+, use 'data_sources' instead of 'sources'
datasource = context.data_sources.add_pandas(name="my_pandas_datasource")
data_asset = datasource.add_dataframe_asset(name="labels_dataset")
```

##6. Define and Create a Data Batch

```python
# Create the definition for the batch using the dataframe 'df' you loaded earlier
batch_definition = data_asset.add_batch_definition_whole_dataframe(name="my_batch_definition")

# Retrieve the batch
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```
